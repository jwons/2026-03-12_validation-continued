```python
import pandas as pd
import numpy as np
import random

# Set seed for reproducibility so you get the same "random" errors every time
np.random.seed(42)
random.seed(42)

# 1. Define dataset size
n_rows = 500

# 2. Generate valid baseline data
patient_ids = [f"P{str(i).zfill(4)}" for i in range(1, n_rows + 1)]
ages = np.random.randint(18, 101, size=n_rows)
arms = np.random.choice(["Placebo", "Drug_A", "Drug_B"], size=n_rows)
scores = np.round(np.random.uniform(0.1, 15.0), size=n_rows, decimals=2)
outcomes = np.random.randint(0, 2, size=n_rows)

df = pd.DataFrame({
    "patient_id": patient_ids,
    "age": ages,
    "treatment_arm": arms,
    "biomarker_score": scores,
    "outcome": outcomes
})

# 3. Select random indices to inject guaranteed errors
# We need 8 distinct indices for our 8 specific error types
error_indices = random.sample(range(n_rows), 8)

# Error 1: Missing patient_id (NaN)
df.loc[error_indices[0], "patient_id"] = np.nan

# Error 2: Duplicate patient_id (copy index 0's ID to this row)
df.loc[error_indices[1], "patient_id"] = df.loc[0, "patient_id"]

# Error 3: Age too low (< 18)
df.loc[error_indices[2], "age"] = 15

# Error 4: Age too high (> 100)
df.loc[error_indices[3], "age"] = 150

# Error 5: Treatment arm case-sensitivity error
df.loc[error_indices[4], "treatment_arm"] = "drug_A"

# Error 6: Treatment arm typo
df.loc[error_indices[5], "treatment_arm"] = "Placbo"

# Error 7: Negative biomarker score
df.loc[error_indices[6], "biomarker_score"] = -4.5

# Error 8: Invalid outcome category
df.loc[error_indices[7], "outcome"] = 3

# 4. Optional: Randomly shuffle the rows so errors aren't just at the end
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

# 5. Export to CSV
file_name = "clinical_trials_batch1.csv"
df.to_csv(file_name, index=False)

print(f"Successfully generated {n_rows} rows and saved to {file_name}.")
print("Guaranteed errors injected for: missing ID, duplicate ID, invalid ages, misspelled treatment arms, negative biomarker, and invalid outcome.")
```
